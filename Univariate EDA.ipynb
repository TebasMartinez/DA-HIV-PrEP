```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("data/filtered_countries.csv")
df = df.drop(columns=["Unnamed: 0"])
df
```

```python
df["country"].unique()
```

# Univariate Analysis

## Categorical Variable

```python
df.columns
```

```python
country_frequency = df["country"].value_counts()
country_proportion = df["country"].value_counts(normalize=True)

country_frequency, country_proportion
```

```python
sns.barplot(x=country_frequency.index, y=country_frequency.values , palette="Set2", hue=country_frequency)
plt.xticks(rotation=90)
plt.show();
```

```python
year_frequency = df["year"].value_counts()
year_proportion = df["year"].value_counts(normalize=True)

country_frequency, country_proportion
```

```python
sns.barplot(x=year_frequency.index, y=year_frequency.values , palette="Set2", hue=year_frequency)
plt.xticks(rotation=90)
plt.show();
```
